# Lineární regrese v R

V naší diskusi statistických metod jsme se zatím ve většině případů omezili na jednu náhodnou veličinu. Zabývali jsme se problémy jako je bodový odhad její střední hodnoty (výpočty intervalu spolehlivosti) a testování nejrůznějších druhů hypotéz o populaci, ze které pochází prvky jednoho nebo více náhodných výběrů, pro které tuto veličinu měříme.

V této kapitole si představíme aparát, který nám umožní zkoumat hypotézu jednoduché lineární závislosti určité náhodné veličiny $y$ na jiné proměnné ($x$) a sestavit lineární regresní model pro tuto závislost. V tomto případě často nazýváme proměnnou $y$ *závisle proměnnou* a $x$ nezávisle proměnnou. Je třeba též zmínit, že $x$ zde neuvažujeme jako náhodnou veličinu a předpokládáme, že ji lze změřit dostatečně přesně, tj. je to pouze veličina $y$, která je zatížena experimentální chybou.

V případě jednoduché lineární regrese přepokládáme, že hledaný vztah mezi závisle proměnnou $y$ a nezávisle proměnnou $x$ existuje následující vztah:

$y = \beta_0 + \beta_1 x $

Jinými slovy lze tento vztah popsat rovnicí přímky, kde parametr $\beta_1$ je její *směrnice* (angl. *slope*) a parametr $\beta_0$ je *úrovňová konstanta* (angl. *intercept*). 

Tento typ vztahu je poměrně častý nebo je minimálně velice dobrou aproximací skutečného děje. Například lze ukázat, že [teplota vzduchu se dá relativně přesně popsat pomocí cvrkání cvrčka](https://en.wikipedia.org/wiki/Dolbear%27s_law) *Oecanthus fultoni* podle rovnice:

$T_{F}=50+\left({\frac  {N_{{60}}-40}{4}}\right) = 40 + \frac{1}{4}N_{{60}}$,

kde $T_{F}$ je teplota ve stupních Fahrenheit (závislá proměnná) a $N_{{60}}$ je počet cvrků cvrčka za minutu (nezávislá proměnná). Tento vztah bychom tedy mohli získat z experimentu, kde budeme měřit cvrkání tohoto cvrčka pro různé teploty a pomocí regresní analýzy pak stanovit koeficienty $\beta_0$ a $\beta_1$. Měli bychom tak získat hodnoty odpovídající $\beta_1 \approx \frac{1}{4}$ a $\beta_0 \approx 40$.

## Modelové příklady

### Definice předpisu lineární funkce

V této lekci si budeme principy lineární regrese ukazovat na modelových příkladech, které si sami vytvoříme. Za tímto účelem si tedy nejprve musíme definovat funkci, která nám vygeneruje modelovaný lineární vztah:

```r
get_linear_relationship <- function(b_0, b_1) {
    print(paste('Generating linear relationship:', b_0, '+', b_1,'* x'))
    
    return(
        function(x) {
            return(b_0 + b_1 * x)
        }
    )
}

ultimate_truth <- get_linear_relationship(5, 2)
ultimate_truth
```

Funkce `get_linear_relationship` nám tedy po specifikaci parametrů $\beta_0$ a $\beta_1$  vrátí novou funkci, která definuje příslušný vztah a můžeme ji tak použít pro výpočet hodnot závisle proměnné $y$ z hodnot nezávisle proměnné $x$:

```r
x <- -10:10
x
ultimate_truth(x)
```

Tuto funkci si můžeme představit jako absolutní pravdu, kterou se snažíme našimi experimenty odhalit. V praxi je nám totiž přesný předpis této funkce skrytý a její parametry $\beta_0$ a $\beta_1$ se snažíme stanovit z dostupných dat.

Příslušnou přímku si můžeme snadno zobrazit pomocí grafu:

```r
library('ggplot2')

my_plot_theme <- function(plt) {
    # zvyrazni osy soustavy souradnic a pouzij minimalisticke tema
    plt + geom_hline(
        yintercept = 0, 
        linetype="dashed"
    ) + geom_vline(
        xintercept = 0,
        linetype="dashed"
    ) + theme_minimal()
}

plot_function <- function(fun, x) {
    plt <- ggplot(
        data.frame(
            x = x,
            y = fun(x)
        ),
        aes(
            x = x,
            y = y
        )
    ) + geom_line(
        color='navyblue'
    )
    
    return(my_plot_theme(plt))
} 

plot_function(ultimate_truth, -10:10)
```

### Simulace měření

Nyní si napíšeme funkci, která bude simulovat měření proměnných $x$ a $y$:

```r
constant_measurements <- function(n_measurements, xmin = -1, xmax = 1) {
    # vrátí simulované hodnoty pro proměnnou x
    # vzdálenost mezi hodnotami x je konstantní a odpovídá počtu měření
    seq(xmin, xmax, (xmax - xmin) / n_measurements)
}

normal_error <- function(y, sd = 1) {
    # aplikuje náhodnou chybu na každou hodnotu y
    # vygenerované chyby mají normální rozdělení
    y + rnorm(length(y), mean = 0, sd = sd)
}

simulate_measurements <- function(
    fun, # funkce definiující zkoumaný vztah
    measuring_strategy, # nastavuje rozsah hodnot proměnné x podle počtu měření
    measuring_error, # nastaví způsob simulace experimentální chyby
    n_measurements = 100 # počet provedených měření
) {
    # použij definovanou strategii pro stanovení hodnot x
    x <- measuring_strategy(n_measurements)
    
    # získej skutečné hodnoty pro závislou proměnnou y
    y <- fun(x)
    
    # simuluj experimentální chybu
    y <- measuring_error(y)
    
    return(data.frame(x=x,y=y))
}

df_experiment <- simulate_measurements(
    ultimate_truth,
    measuring_strategy = constant_measurements,
    measuring_error = normal_error,
    n_measurements = 10
)
dim(df_experiment)
df_experiment
```

Zavoláním funkce `simulate_measurements` tedy simulujeme průběh měření veličin $x$ a $y$. Počítáme zde s určitým způsobem volby hodnot $x$, která většinou záleží na experimentálním uspořádání. My jsme zde zvolili strategii, kde si interval proměnné $x$ rozdělíme na stejně velké části podle počtu požadovaných měření. V praxi by to mohli být například stejně velké vzdálenosti od zdroje záření, postupně zvětšované teploty o 1°C atp. Přesnou hodnotu pro veličinu $y$ pak spočítáme pomocí dříve definovaného lineárního vztahu. Jelikož ale svět není ideální, zatížíme tyto skutečné hodnoty simulovanou experimentální chybou, která v tomto případě pochází z normálního rozdělení.

Náš simulovaný experiment si opět můžeme ukázat v grafu:

```r
plot_experiment <- function(data, fun = NULL) {
    plt <- ggplot(
        data,
        aes(
            x = x,
            y = y
        )
    ) + geom_point(
        color='navyblue'
    )
    
    if(!is.null(fun)) {
        plt <- plt + geom_line(
            aes(
                y=fun(x)
            ),
            color='darkred'
        )
    }
    
    return(my_plot_theme(plt))
}

plot_experiment(df_experiment, ultimate_truth)
```

Modré body zde odpovídají simulovaným měřením veličin $x$ a $y$. Původní funkce, jejíž parametry se snažíme nalézt, je vykreslena jako červená přímka.

Máme tedy vše potřebné pro simulaci nejrůznějších situací, které by mohli v praxi nastat při experimentálním stanovení. Zkuste si schválně pomocí kódu níže vygenerovat různé situace. Zaměřte se především na rozdělení nezávislé veličiny $x$ (`measuring_strategy`), volbu rozptylu náhodné chyby měření (`sd`) a počtu vzorků (`n_measurements`):

```r
ultimate_truth <- get_linear_relationship(b_0 = 5, b_1 = 2)

df_experiment <- simulate_measurements(
    ultimate_truth,
    measuring_strategy = function(n_measurements) {
        constant_measurements(n_measurements, xmin = -10, xmax = 10)
    },
#     measuring_strategy = rnorm,
#     measuring_strategy = rlnorm,
#     measuring_strategy = runif,
    measuring_error = function(y) {
        normal_error(y, sd = 2)
    },
    n_measurements = 100
)
plot_experiment(df_experiment, ultimate_truth)
```

## Pearsonův korelační koeficient

Představme si nyní, že neznáme tvar funkce `ultimate_truth` a nevíme ani to, že vůbec existuje lineární závislost proměnné $y$ na $x$. Zkrátka dostali jsme do rukou následující data:

```r
dim(df_experiment)
head(df_experiment)
```

a je na nás zjistit, zda existuje závislost mezi $y$ a $x$ a tuto závislost co nejpřesněji popsat lineárním regresním modelem, pokud je to možné.

Samozřejmě si můžeme obě proměnné zobrazit graficky a pohledem posoudit, zda mají naše data lineární tendenci:

```r
plot_experiment(df_experiment)
```

To je často velice dobrý způsob jak začít, ale lepším pomocníkem je tzv. Pearsonův korelační koefficient, který číselně kvantifikuje míru lineární závislosti mezi dvěma náhodnými proměnnými $X$ a $Y$:

$\rho _{X,Y}={\frac {\operatorname {cov} (X,Y)}{\sigma _{X}\sigma _{Y}}}$

Tento koeficient nabývá hodnoty v intervalu od -1 do +1. Jedná se v podstatě o jakousi normalizovanou hodnotu [kovariance](https://en.wikipedia.org/wiki/Covariance) mezi $X$ a $Y$, což je ve zkratce jakýsi vzájemný rozptyl dvou náhodných veličin $X$ a $Y$. Rozptyl jedné náhodné veličiny ($\sigma$) je totiž pouze speciální případ pro kovarianci veličiny $X$ se sebou samou:

$ \operatorname {cov} (X,X)=\operatorname {var} (X)\equiv \sigma ^{2}(X)\equiv \sigma _{X}^{2}$

Výběrovou kovarianci dvou výběrů, můžeme v R snadno spočítat pomocí funkce `cov`: 

```r
cov_xy <- cov(x = df_experiment$x, y = df_experiment$y)
cov_xy
```

Samotný Pearsonův korelační koeficient nám pak podle vzorečku nahoře vyjde takto:

```r
cov_xy / (sd(df_experiment$x) * sd(df_experiment$y))
```

Samozřejmě ale v R existuje i samostatná funkce:

```r
cor(x = df_experiment$x, y = df_experiment$y, method = 'pearson')
```

Hodnota korelačního koeficientu -1 znamená perfektní negativní závislost. Tedy závislost, kde by směrnice přímky měla negativní hodnotu:

```r
plot_function(get_linear_relationship(b_0 = 5, b_1 = -2), -10:10)
```

Naopak hodnota +1 by znamenala, že všechny body našeho souboru leží na přímce s pozitivní směrnicí:

```r
plot_function(get_linear_relationship(b_0 = 5, b_1 = 2), -10:10)
```

Hodnota 0 znamená, že neexistuje lineární závislost mezi veličinou $X$ a $Y$. Směrnice přímky by byla rovná 0, a tudíž lze považovat obě veličiny za nezávislé a lineární regrese by neměla smysl (hodnota $y$ se nijak nemění s $x$, protože $\beta_1 = 0$). 

Obecně platí, že pokud je Pearsonův korelační koeficient v absolutní hodnotě větší nebo roven 0.5 lze hovořit o lineární závislosti.

### Test nezávislosti o korelačním koeficientu

Se stanovením případné lineární závislosti nám též pomůže i statistický test, který testuje alternativní hypotézu, která říká, že korelační koeficient je nenulový:

```r
cor.test(df_experiment$x, df_experiment$y)
```

Pokud se nám tedy podaří zamítnout nulovou hypotézu, můžeme s relativní jistotou tvrdit, že naše data jsou lineárně korelovaná. V našem simulovaném případě se toto skutečně potvrdilo s poměrně velkou mírou spolehlivosti.

## Konstrukce regresní přímky

Běžná regresní přímka se konstruuje pomocí minimalizace součtu čtverců vzdáleností jednotlivých bodů od této přímky. Pokud tedy pro přímku předpokládáme vztah:

$y = \beta_0 + \beta_1 x $

hledáme takové koeficienty $\beta_0$ a $\beta_1$, které minimalizují sumu čtverců:

$S(\beta_0,\beta_1)=\sum _{i=1}^{n}{[y(x_{i})-y_{i}]^{2}}=\sum _{i=1}^{n}{(\beta_0 + \beta_1x_{i}-y_{i})^{2}}$,

kde $y(x_{i})$ je hodnota hledané regresní funkce $y$ pro hodnotu $x_{i}$ v $i$-tém vzorku a $y_{i}$ je experimentálně naměřená hodnota závislé proměnné pro vzorek $i$. Jednotlivé členy $[y(x_{i})-y_{i}]^{2}$ si pak opravdu můžeme představit jako čtverce a $S(\beta_0,\beta_1)$ je pak suma jejich ploch: 

<img src="https://spot.pcc.edu/math/ahss/ed2/images/regression/leastSquares.png" width="500px" />

Abychom nalezli minimum $S(\beta_0,\beta_1)$, musíme tedy obě parciální derivace této funkce položit rovny nule:

${\partial S \over \partial \beta_1}=2\sum _{i=1}^{n}(\beta_0 + \beta_1x_{i}-y_{i})x_{i}=0$

${\partial S \over \partial \beta_0}=2\sum _{i=1}^{n}(\beta_0 + \beta_1x_{i}-y_{i})=0$

Úpravami těchto rovnic lze získat předpisy pro oba koeficienty, ale tím se zde již zabývat nebudeme. Lze též ukázat, že existuje pouze jedno řešení, a tudíž pomocí této metody vždy nalezneme optimální rovnici, která odpovídá minimu $S(\beta_0,\beta_1)$. 

Linární regresi lze v R spočítat pomocí funkce `lm` (linear model), která udělá tento výpočet za nás:

```r
lm_experiment <- lm(df_experiment$y ~ df_experiment$x)
lm_experiment
```

Přesné hodnoty odhadů koeficientů $\beta_0$ a $\beta_1$ pak z modelu získáme pomocí příslušného atributu:

```r
lm_experiment$coefficients
```

Je dobré si uvědomit, že tyto hodnoty jsou pouze odhady a jsou tudíž též náhodnými veličinami zatíženými experimentální chybou. Pokud bychom totiž ten samý experiment provedli několikrát, získáme různé odhady těchto parametrů. Zkusme si vytvořit jednoduchou simulaci:

```r
simulate_measurements_multiple_experiments <- function(
    measuring_strategy,
    measuring_error,
    n_measurements = 100,
    n_experiments = 10 # počet opakování experimentu
) {
    df_experiment <- NULL
    for (sample_id in 1:n_experiments) {
        # opakování experimentu v každé iteraci cyklu
        df_sample <- simulate_measurements(
            ultimate_truth,
            measuring_strategy = measuring_strategy,
            measuring_error = measuring_error,
            n_measurements = n_measurements
        )
        df_sample$ExperimentID <- paste('Experiment', sample_id)

        df_experiment <- rbind(df_experiment, df_sample)
    }
    
    return(df_experiment)
}

df_experiments <- simulate_measurements_multiple_experiments(
    measuring_strategy = constant_measurements,
    measuring_error = normal_error,
    n_measurements = 10, # deset měření na jeden experiment 
    n_experiments = 10 # deset experimentů 
)
dim(df_experiments)
head(df_experiments, 22)
```

Vytvořili jsme si tedy deset různých experimentů, kde v každém z nich pro deset předem daných hodnot $x$ stanovíme příslušnou hodnotu $y$:

```r
plot_experiment(df_experiments)
```

Data vztahující se ke konkrétnímu experimentu si můžeme zvýraznit jednoduchou modifikací funkce `plot_experiment`:

```r
plot_experiment <- function(data, sampleCol = NULL) {    
    plt <- ggplot(
        data,
        aes(
            x = x,
            y = y
        )
    )
    
    if (is.null(sampleCol)) {
        plt <- plt + geom_point(
            color = 'navyblue'
        )
    } else {
        plt <- plt + geom_point(
            aes_string(
                color=sampleCol
            )
        )
    }
    
    return(my_plot_theme(plt))
}

df_experiments$ExperimentID <- as.factor(df_experiments$ExperimentID)
plot_experiment(df_experiments, 'ExperimentID')
```

Získáme tedy logicky deset různých odhadů parametrů regresního modelu: 

```r
lm_experiments <- by(
    df_experiments, 
    df_experiments$ExperimentID, 
    function(experiment) {
        lmodel <- lm(experiment$y ~ experiment$x)
        names(lmodel$coefficients) <- NULL
        list(
            b_0 = lmodel$coefficients[1],
            b_1 = lmodel$coefficients[2],
            ExperimentID = experiment$ExperimentID[1]
        )
    }
)

do.call(rbind, lm_experiments)
```

Pro každý bod $x$ v našem souboru pak získáme různou predikci podle modelu pro daný experiment:

```r
use_model <- function(b_0, b_1, x) {
    b_0 + b_1 * x
}

df_experiments$y_model <- unlist(lapply(
    unique(df_experiments$ExperimentID), 
    function(measurement) {
        model <- lm_experiments[[measurement]]
        
        mask <- df_experiments$ExperimentID == measurement
        return(
            use_model(
                model$b_0, 
                model$b_1, 
                df_experiments[mask,'x']
            )
        )
    }
))

head(df_experiments)
```

Vypočítané predikce pak můžeme využít pro vizualizaci jednotlivých regresních přímek:

```r
plot_experiment(df_experiments, 'ExperimentID') + geom_line(
    aes_string(
        y='y_model',
        color='ExperimentID'
    )
)
```

Dostáváme tedy deset různých regresních přímek, kde každá z nich má různý tvar podle parametrů $\beta_0$ a $\beta_1$.

### Intervaly spolehlivosti pro parametry regresní přímky

#### Rezidua

Do tohoto grafu si též můžeme snadno zobrazit tzv. *rezidua*, což jsou vlastně strany pomyslných čtverců, které jsme si [popsali výše při popisu konstrukce regresní přímky](10.ipynb#Konstrukce-regresn%C3%AD-p%C5%99%C3%ADmky):

```r
plt + geom_point(
    # body na regresní přímce pro vstup x
    aes_string(
        y='y_model'
    ),
    color='darkred'
) + geom_segment(
    # rezidua
    aes_string( 
        xend='x', 
        yend='y_model'
    ),
    color='darkcyan'
)
```

Rezidua jsou důležitou vlastností regresního modelu, neboť nám poskytují odhad náhodných chyb a lze je snadno spočítat:

```r
residuals <- df_experiment$y - df_experiment$y_model
sd(residuals)
mean(residuals)
```

Vidíme, že naše rezidua mají průměr skoro rovný 0 se směrodatnou odchylkou cca 2. To není náhoda. Vzpomeňte si, že náhodné chyby jsme si v naší simulaci zvolili z normálního rozdělení s průměrem 0 a směrodatnou odchylkou 2. Zkuste si schválně zvolit jinou distribuci chyb a podívat se, jak se nám změní výstup celé naší simulace:

```r
ultimate_truth <- get_linear_relationship(b_0 = 5, b_1 = 2)

df_experiment <- simulate_measurements(
    ultimate_truth,
    measuring_strategy = function(n_measurements) {
        constant_measurements(n_measurements, xmin = -10, xmax = 10)
    },
    measuring_error = function(y) {
        normal_error(y, sd = 2) # zmena zde
    },
    n_measurements = 100 # jak se zmeni odhad smerodatne odchylky rezidui pro vetsi pocet bodu?
)

lm_experiment <- lm(df_experiment$y ~ df_experiment$x)

df_experiment$y_model <- use_model(
    lm_experiment$coefficients[1], 
    lm_experiment$coefficients[2], 
    df_experiment$x
)

residuals <- df_experiment$y - df_experiment$y_model
sd(residuals)
mean(residuals)
```

Je též užitečné si uvědomit, že v tomto ideálním případě, kde experimentální chyba má normální rozdělení, je suma reziduí přibližně rovná 0:

```r
sum(residuals)
```

#### Reziduální rozptyl

V souvislosti s lineární regresí se též uvádí i tzv. *reziduální rozptyl* $s^2$ (*residual standard error*), který je nestranným odhadem rozptylu náhodné veličiny $y$ podél naší regresní přímky:

$s^2 = \frac{1}{n-2} SS_R$, 

kde $SS_R$ je tzv. *reziduální součet čtverců* (*residual sum of squares*):

$SS_R = \sum_{j=1}^{n}{e_j^2}$,

kde $e_j$ je reziduum příslušející vzorku $j$. 

V R bychom si reziduální rozptyl spočátali takto:

```r
SS_R <- sum(residuals ** 2)
sqrt(
    SS_R / (length(residuals) - 2)
)
```

Člen $SS_R$ odpovídá tedy součtu ploch čtverců, které minimalizujeme při konstrukci regresní přímky a dělíme jej počtem reziduí, od kterého odečteme číslo 2. Proč ale? 

Číslo dvě je totiž důvodem, proč je tento odhad lepší než pouhá směrodatná odchylka reziduí jako jsme si ukázali výše:

```r
sd(residuals)
```

Jmenovatel výrazu totiž odpovídá stupňům volnosti našeho modelu. Počet stupňů volnosti je roven počtu datových bodů, od kterého odečteme počet odhadovaných parametrů modelu, což je dva ($\beta_0$ a $\beta_1$). Pokud bychom odečetli pouze jedničku, máme totiž přesně vzorec pro výpočet směrodatné odchylky reziduí:

```r
DF <- length(residuals) - 2
DF
```

```r
sd(residuals)
sqrt(
    SS_R / (length(residuals) - 1)
)
```

### Koeficient determinace

Krom reziduálního rozptylu popsaného výše je další často užívanou statistikou při odhadu chyby našeho modelu tzv. *koeficient detrminace* (*coefficient of determination*, *R-squared* nebo jen *R2*): 

```r
summary(lm_experiment)
```

Vidíme jej skoro úplně na konci našeho shrnutí jako `Multiple R-squared`. 

Koeficient determinace se spočte podle následujícího vzorce:

$R^{2}=1-{SS_R \over SS_T}$, 

kde $SS_R$ je reziduální suma čtverců tak, jak jsme si ji [uvedli výše](./10.ipynb#Rezidu%C3%A1ln%C3%AD-rozptyl) a $SS_T$ je tzv. *total sum of squares*, která se vypočte jako:

$SS_{T}=\sum_{j=1}^n (y_{j} - \overline{y})^2$,

kde $y_{j}$ je hodnota veličiny $y$ pro měření $j$ a $\overline{y}$ je výběrový průměr veličiny $y$. Koeficient determinace bychom tedy v R získali takto:

```r
SS_R <- sum(lm_experiment$residuals ** 2)
SS_T <- sum((df_experiment$y - mean(df_experiment$y)) ** 2)
1 - (SS_R / SS_T)
```

```r
summary(lm_experiment)$r.squared
```

Ve výstupu funkce `summary` jste si možná ještě všimli výstupu označeného jako `Adjusted R-squared`. Jelikož lze lineární regresi použít i pro problémy, kde máme více nezávislých proměnných $x$, vzniká u koeficientu determinace problém. Tento koeficient totiž roste s počtem proměnných, a tudíž není dostatečně nestranný. Vícerozměrnou lineární regresi si však necháme na jindy, a proto tuto hodnotu zatím můžeme ignorovat. Měla by totiž pro jednu proměnnou být velice blízko běžnému koeficientu determinace.

```r
summary(lm_experiment)$adj.r.squared
```

### Předpoklad homoscedasticity dat v lineární regresi

Stejně jako u anlýzy rozptylu je i u lineární regrese důležitá homoscedasticita dat, tedy rovnost směrodatných odchylek reziduí pro všechny hodnoty proměnné $x$ (u analýzy rozptylu to byla směrodatná odchylka jednotlivých vzorků). Homoscedasticitu dat bychom si třeba mohli uměle narušit v naší simulaci. Definujme si například následující funkci pro generování chyby našeho měření:

```r
nonhomeoscedastic_error <- function(y) {
    sd <- 2
    sapply(y, function(val) {
        sd <<- sd * 1.03 # všimněte si využití operátoru '<<-' pro zapsání do proměnné mimo prostředí funkce
        val + rnorm(1, 0, sd)
    })
}
```

Zde využijeme faktu, že naše data jsou již seřazená a chybu zvětšujeme postupně o 3% podél našeho datového souboru. Tato chybová funkce nám zaručí, že pro větší hodnoty x, budeme mít větší chybu. Efekt takto generované chyby je jasný z následující vizualizace:

```r
ultimate_truth <- get_linear_relationship(5, 2)

df_experiment <- simulate_measurements(
    ultimate_truth,
    measuring_strategy = function(n_measurements) {
        constant_measurements(n_measurements, xmin = -10, xmax = 10)
    },
    measuring_error = nonhomeoscedastic_error, # naše nová error funkce -> způsobuje nestabilitu a nesprávný odhad parametrů
#     measuring_error = function(y) { # původní error funkce
#         normal_error(y, sd = 2)
#     },
    n_measurements = 100
)

lm_experiment <- lm(df_experiment$y ~ df_experiment$x)
lm_experiment$coefficients
plot_experiment(df_experiment) + geom_smooth(
    method=lm,
    formula=y~x,
    color="darkred", 
    fill="darkcyan", 
    se=TRUE,
    level=0.99
)
```

Pokud si simulaci výše zkusíte spustit několikrát za sebou, zjistíte, že odhad skutečných parametrů naší přímky je teď mnohem měně stabilní a často dostáváme velice různé hodnoty, než jsou ty skutečné. Porovnejte si to schválně s variabilitou našeho modelu při použití funkce `normal_error`. I když zvětšíte rozptyl chyb, bude se odhad parametrů pořád měnit mnohem méně. Interval spolehlivosti pro parametry našeho modelu zde též nedává moc smysl (pás intervalu spolehlivosti evidentně vůbec nereflektuje skutečnou distribuci naší chyby podel regresního modelu).


### Shrnutí předpokladů lineární regrese

Předpoklady využití lieární regrese jsme si všechny již vyjmenovali, ale pojďme si ještě všechno shrnout na jedno místo.

1. **Modelovaný vztah musí mít lineární charakter.** To znamená, že závislá proměnná $y$ je s nezávislou proměnnou $x$ spjatá lineárně. Lze ji tedy vyjádřit vztahem $y = \beta_0 + \beta_1 x $.
2. **Rezidua musí být nezávislá a musí mít identickou a to normální distribuci.** Rezidua tedy musí být náhodnou proměnnou, která pochází z jednoho normálního rozdělení.
3. **Rezidua musí mít konstantní rozptyl.** Je tedy nutné aby data byla homeoscedastická.

Toto jsou tedy tři základní a jediné předpoklady pro provedení dostatečně robustní lineární regrese. Všimněte si, že vůbec není nutné aby proměnná $x$ nebo $y$ byla normálně rozdělená. V naší simulaci byla například proměnná $x$ většinou rozdělená nenormálně, viz histogram:

```r
hist(df_experiment$x)
```

Poměrně přesné výsledky bychom totiž zjistili i v případě jiného rozdělení. Například pro lognormální rozdělení by naše regrese vypadala takto:

```r
ultimate_truth <- get_linear_relationship(5, 2)

df_experiment <- simulate_measurements(
    ultimate_truth,
    measuring_strategy = rlnorm,
    measuring_error = function(y) {
        normal_error(y, sd = 2)
    },
    n_measurements = 100
)

hist(df_experiment$x)
hist(df_experiment$y)
```

Pokud se podívate na odhad parametrů, není to vůbec špatné a v podstatě to odpovídá výsledkům našich předchozích experimentů:

```r
lm_experiment <- lm(df_experiment$y ~ df_experiment$x)
lm_experiment$coefficients
plot_experiment(df_experiment) + geom_smooth(
    method=lm,
    formula=y~x,
    color="darkred", 
    fill="darkcyan", 
    se=TRUE,
    level=0.99
)
```

```r
plot_experiment(
    data.frame(
        x = df_televisions$PeoplePerPhysician,
        y = df_televisions$LifeExpectancy
    )
)
```

Evidentně se nejedná o normálně rozdělená data, ale pomůže nám zde logaritmizace osy x:

```r
plot_experiment(
    data.frame(
        x = df_televisions$PeoplePerPhysician,
        y = df_televisions$LifeExpectancy
    )
) + scale_x_log10()
```

Zdá se tedy, že existuje mírná záporná lineární závislost mezi počtem lidí na jednoho doktora. To je logické, neboť čím více lidí na jednoho doktora připadne, tím horší pravděpodobně bude zdravotní péče a tím se sníží i délka života. 

Můžeme si samozřejmě vytvořit i příslušnou regresní přímku a zjistit tak, jak konkrétně střední délka života závisí na dostupnosti lékařské péče:

```r
lm_televisions <- lm(df_televisions$LifeExpectancy ~ df_televisions$PeoplePerPhysician)
lm_televisions$coefficients
plot_experiment(
    data.frame(
        x = df_televisions$PeoplePerPhysician,
        y = df_televisions$LifeExpectancy
    )
) + geom_smooth(
    method=lm,
    formula=y~x,
    color="darkred", 
    fill="darkcyan", 
    se=TRUE,
    level=0.99
) + scale_x_log10()
```

```r
summary(lm_televisions)
```

Jak to ale bude v případě proměnné *PeoplePerTV*. Intuitivně bychom řekli, že dostupnost k televiznímu přístroji pravděpodobně zdraví nepomůže, ale co na to náš model? 

```r
cor(df_televisions$PeoplePerTV, df_televisions$LifeExpectancy)
```

Kupodivu zde korelace existuje a dokonce silnější, než u proměnné *PeoplePerPhysician*! Podívejme se na příslušný regresnní model:

```r
lm_televisions <- lm(df_televisions$LifeExpectancy ~ df_televisions$PeoplePerTV)
lm_televisions$coefficients
plot_experiment(
    data.frame(
        x = df_televisions$PeoplePerTV,
        y = df_televisions$LifeExpectancy
    )
) + geom_smooth(
    method=lm,
    formula=y~x,
    color="darkred", 
    fill="darkcyan", 
    se=TRUE,
    level=0.99
) + scale_x_log10()
```

Dostali jsme velice podobnou závislost. Jak je to ale možné? Jsou opravdu televize stejně důležité jako doktoři? Pravděpodobě ne...

Zde je třeba si vzpomenout na poučku, že korelace není kauzalita ([Correlation does not imply causation](https://en.wikipedia.org/wiki/Correlation_does_not_imply_causation)). Je velice pravděpodobné, že zde existuje třetí proměnná, tzv. *zavádějící faktor* (angl. *confounding variable*). Tato proměnná ovlivňuje jak počet doktorů v dané zemi, tak i počet televizí. Dostáváme tedy správně statisticky významnou korelaci, ale příčina je úplně jiná. Zavádějícím faktorem by v tomto případě mohl být hrubý domácí produkt dané země. Čím silnější je ekonomika dané země, tím více doktorů a tím pádem i televizí si bude moci dovolit.

Na tomto příkladu je tedy vidět, že i když dokážeme najít spolehlivou korelaci mezi dvěma proměnnými, nemusí to vůbec znamenat, že jsou to právě tyto proměnné, které mají skutečný vliv na pozorovaný efekt. Při interpretaci našich modelů bychom tedy měli vždy být maximálně opatrní.
