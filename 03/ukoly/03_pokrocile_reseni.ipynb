## Počítání písmen

```r
# kratší verze s horším formátováním

text_stats <- function(text) {
    text <- unlist(strsplit(tolower(text), ""))
    
    char_counts <- integer(length(letters))
    for (char in text) {
        if (char %in% letters) {
            char_counts[which(char == letters)] <- char_counts[which(char == letters)] + 1
        }
    }
    
    print(letters[char_counts > 0])
    print(char_counts[char_counts > 0])
}
```

Testovací kód:

```r
text_stats("Hi! I hope noone is reading this.")
```

```r
# delší verze s lepším formátováním

text_stats <- function(text) {
    text <- unlist(strsplit(tolower(text), ""))
    
    char_counts <- integer(length(letters))
    for (char in text) {
        if (char %in% letters) {
            char_counts[which(char == letters)] <- char_counts[which(char == letters)] + 1
        }
    }
    
    for (idx in 1:length(char_counts)) {
        if (char_counts[idx] != 0) {
            print(paste(
                letters[idx]
                , char_counts[idx]
            ))
        }
    }
}
```

Testovací kód:

```r
text_stats("Hi! I hope noone is reading this.")
```

## Šifra

```r
encrypt <- function(text) {
    text <- unlist(strsplit(tolower(text), ""))
    key <- c(letters, c(" ", ".", "!", "?", ","))
    
    text_encrypted <- c()
    for (char in text) {
        text_encrypted <- c(text_encrypted, which(char == key))
    }
    
    text_encrypted
}

decrypt <- function(cypher) {
    key <- c(letters, c(" ", ".", "!", "?", ","))
    
    text_decrypted <- c()
    for (code in cypher) {
        text_decrypted <- c(text_decrypted, key[code])
    }
    
    paste(text_decrypted, sep = "", collapse="")
}
```

Testovací kód:

```r
encrypted_text <- encrypt(
    "Hi! I hope noone is reading this."
)
encrypted_text
decrypt(encrypted_text)
```

## Extrakce slov

```r
# asi ideální řešení
extract_words <- function(text, unique = F) {
    text <- gsub("[,!\\.?]", "", tolower(text))
    text <- unlist(strsplit(text, " "))
    
    if (unique) {
        # všimněte si, že zde nevadí když máme ve funkci lokální proměnnou, 
        # která se jmenuje stejně jako interní funkce unique, R si s tím poradí
        # (většinou se ale podobným praktikám vyhýbejte a snažte se vždy proměnné označovat unikátně)
        text <- sort(unique(
            text
        ))
    }
    
    text
}
```

Testovací kód:

```r
extract_words("Ahoj, jak se máš?")
extract_words("Ahoj, jak se máš? Tahle funkce se ti fakt povedla!")
extract_words("Ahoj, jak se máš? Já se mám dobře. Tak zase někdy. Ahoj.")
extract_words("Ahoj, jak se máš? Já se mám dobře. Tak zase někdy. Ahoj.", unique = T)
```

```r
# řešení bez funkce unique a regulárních výrazů

extract_words <- function(text, unique = F) {
    text <- unlist(
        strsplit(
            tolower(text)
            , " "
        )
    )
    
    text_clean_vect <- c()
    for (word in text) {
        chars <- unlist(strsplit(word, ""))
        word_clean <- c()
        for (char in chars) {
            if (!(char %in% c("!", ".", "?", ","))) {
                word_clean <- c(word_clean, char)
            }
        }
        word_clean <- paste(word_clean, sep = "", collapse="")
        if (unique && (word_clean %in% text_clean_vect)) {
            next()
        }
        text_clean_vect <- c(text_clean_vect, word_clean)
    }
    
    if (unique) {
        text_clean_vect <- sort(text_clean_vect)
    }
    
    text_clean_vect
}
```

Testovací kód:

```r
extract_words("Ahoj, jak se máš?")
extract_words("Ahoj, jak se máš? Tahle funkce se ti fakt povedla!")
extract_words("Ahoj, jak se máš? Já se mám dobře. Tak zase někdy. Ahoj.")
extract_words("Ahoj, jak se máš? Já se mám dobře. Tak zase někdy. Ahoj.", unique = T)
```

## Apply challenge

```r
my_apply <- function(items, func) {
    func(items[1])
    if (length(items[-1])) {
        items <- items[-1]
        my_apply(items, func)
    }
}
```

Testovací kód:

```r
my_apply(1:10, print)
my_apply(1:10, function(x){print(as.character(x))})
```
