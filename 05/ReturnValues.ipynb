```r
df_ufocase <- read.table('ufocase.csv', sep=',', header=T)
head(df_ufocase)
```

```r
df_vector <- apply(df_ufocase, 1, function(x) {
    return(
        x[1] # Year
    )
})
df_vector
```

```r
is.vector(df_vector)
```

```r
df_vector <- apply(df_ufocase, 1, function(x) {
    return(
        c(
            x[1], # Year
            x[2], # Name
            x[3] # Location
        )
    )
})
df_vector
```

```r
class(df_vector)
```

```r
dim(df_vector)
```

```r
is.vector(df_vector)
```

```r
as.data.frame(t(df_vector))
```

```r
df_vector <- apply(df_ufocase, 1, function(x) {
    return(
        matrix(c(
            x[1], # Year
            x[2], # Name
            x[3], # Location
            1
        ), nrow=2, ncol=2)
    )
})
df_vector
```

```r
class(df_vector)
```
