```r
df_imdb <- read.table('ukoly/imdb.csv', sep=';', header=T)
df_imdb$var1 <- NULL
head(df_imdb)
```

```r
df_imdb_year_split <- split(
    df_imdb,
    df_imdb$year
)

head(df_imdb_year_split)
```

```r
fx <- function(x) {
    # jakoze komplexni a zdlouhava funkce
    
    Sys.sleep(0.1) # tohle by treba mohlo byt trenovani neuronove site
          
    return(list( # a toto vystupni statistiky
        mean=mean(x$rating),
        max=max(x$rating),
        min=min(x$rating)
    ))
}
```

```r
system.time(
  results <- lapply(
      df_imdb_year_split,
      fx
  )
)

results
```

```r
library(parallel)

system.time(
  results <- mclapply(df_imdb_year_split, fx, mc.cores = 5)
)

results
```
