## GenreCount

Přidejte do tabulky nový sloupec `GenreCount`, který bude označovat, do kolika žánrů najednou film patří.

```r
genres <- c(
    'Short', 
    'Romance', 
    'Documentary', 
    'Drama', 
    'Comedy', 
    'Animation', 
    'Action'
)
```

```r
# rychle s vyuzitim jen latky z tutorialu

ptm <- proc.time()

genre_count <- rep(0, nrow(df_imdb))
for (genre in genres) {
    genre_count <- genre_count + df_imdb[[genre]]
}
df_imdb[['GenreCount']] <- genre_count

print(proc.time() - ptm)

head(df_imdb)
```

```r
# nebo o fous rychlejsi podle Katky Liskove pomoci funkce rowSums

ptm <- proc.time()

df_imdb["GenreCount"] <- rowSums(df_imdb[genres])

print(proc.time() - ptm)

head(df_imdb)
```

```r
# mozne i reseni pomoci apply ale je trochu pomalejsi

ptm <- proc.time()

df_imdb[['GenreCount']] <- apply(
    df_imdb[genres],
    1,
    sum 
)

print(proc.time() - ptm)
                                 
head(df_imdb)
```

```r
# takhle spis ne 
# (cykly jsou vetsinou extremne pomale pro vetsi data)

ptm <- proc.time()

for (row_idx in 1:nrow(df_imdb)) {
    df_imdb[row_idx,'GenreCount'] <- sum(df_imdb[row_idx,genres])
}

print(proc.time() - ptm)

head(df_imdb)
```

## Top Three Per Genre

Zobrazte tři nejlépe hodnocené filmy pro každý žánr. Uložte jako seznam (`list`), kde klíč je název žánru a hodnota je vektor jmen těchto filmů.

**Bonus:** Omezte se pouze na filmy s alespoň 5,000 hodnoceními (`votes`) a přidejte jejich hodnocení (`rating`) do výstupu, který teď bude seznam (`list`) objektů `data.frame`.

```r
best_of_genres <- list()
for (genre in genres) {
    mask <- df_imdb[[genre]] == 1
    genre_subset <- df_imdb[mask,]
    best_of_genres[[genre]] <- as.vector(
        genre_subset[order(genre_subset$rating, decreasing = T),'title'][1:3]
    )
}
best_of_genres
```

```r
# bonus

est_of_genres <- list()
for (genre in genres) {
    mask <- df_imdb[[genre]] == 1 & df_imdb$votes > 5000
    genre_subset <- df_imdb[mask,]
    best_of_genres[[genre]] <- genre_subset[order(genre_subset$rating, decreasing = T),c('title', 'rating')][1:3,]
}
best_of_genres
```

## Genre Averages

Spočítejte průměrná hodnocení filmů v jednotlivých žánrech. Jsou komedie v průměru lépe hodnocené než akční filmy? Zkusíte navrhnout a provést statistický test?

```r
ptm <- proc.time()

genre_averages <- list()
for (genre in genres) {
    mask <- df_imdb[[genre]] == 1
    genre_subset <- df_imdb[mask,]
    genre_averages[[genre]] <- round(mean(genre_subset$rating), 2)
}

print(proc.time() - ptm)

genre_averages
```

```r
ptm <- proc.time()

genre_averages <- as.list(apply(
    df_imdb[genres],
    2,
    function (genre_col) {
        mean(df_imdb[genre_col == 1,]$rating)
    }
))

print(proc.time() - ptm)

genre_averages
```

```r
genre_averages$Comedy > genre_averages$Action
```

```r
genre_data <- list()
for (genre in genres) {
    mask <- df_imdb[[genre]] == 1
    genre_data[[genre]] <- df_imdb[mask,]$rating
}
t.test(genre_data$Comedy, genre_data$Action, alternative = c('greater'))
```

## Romantic Comedy by Decades

Najděte tři nejlepší romantické komedie z každého desetiletí. Filmy v konečném pořadí musí mít alespoň 30 hodnocení.

**Bonus:** Upravte svůj kód tak, aby se dal aplikovat na jakýkoliv žánr (nebo mix žánrů). Vytvořte si funkci s patřičnými parametry.

```r
ptm <- proc.time()

df_imdb['decade'] <- df_imdb$year - df_imdb$year %% 10
best_romcoms <- by(
    df_imdb,
    df_imdb$decade,
    function(imdb_decade) {
        mask <- imdb_decade$Comedy == 1 & imdb_decade$Romance == 1 & imdb_decade$votes >= 30
        romcoms <- imdb_decade[mask,]
        rating_order <- order(romcoms$rating, decreasing = T)
        return(romcoms[rating_order,][1:3,])
    }
)

print(proc.time() - ptm)

best_romcoms[names(best_romcoms)]
```

```r
ptm <- proc.time()

mask <- df_imdb$Comedy == 1 & df_imdb$Romance == 1 & df_imdb$votes >= 30
romcoms <- df_imdb[mask,]
romcoms['decade'] <- romcoms$year - romcoms$year %% 10
best_romcoms <- by(
    romcoms,
    romcoms$decade,
    function(imdb_decade) {
        rating_order <- order(imdb_decade$rating, decreasing = T)
        return(imdb_decade[rating_order,][1:3,])
    }
)

print(proc.time() - ptm)

best_romcoms[names(best_romcoms)]
```

```r
t(sapply(best_romcoms, function(x) paste(x$title, x$rating)))
```

```r
get_best_movies_by_decade <- function (data, genres, top_max = 3, min_votes = 30, time_period = 10) {
    mask <- !logical(dim(data)[1])
    for (genre in genres) {
        mask <- mask & data[[genre]] == 1
    }
    data <- data[mask & data$votes >= min_votes,]
    data['decade'] <- data$year - data$year %% time_period
    
    best <- by(
        data,
        data$decade,
        function(imdb_decade) {
            rating_order <- order(imdb_decade$rating, decreasing = T)
            return(imdb_decade[rating_order,][1:top_max,])
        }
    )
    
    return(t(sapply(best, function(x) paste(x$title, x$rating))))
}

get_best_movies_by_decade(df_imdb, c('Romance', 'Comedy'), 3)
```
